### Soluation step1. Stationarity (ADF Test):

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series, name=''):
    result = adfuller(series.dropna())
    print(f"ADF Test for {name}")
    print(' ADF Statistic:', result[0])
    print(' p-value:', result[1])
    print(' Critical Values:')
    for key, value in result[4].items():
        print(f'    {key}: {value}')
    if result[1] < 0.05:
        print(" => Stationary\n")
    else:
        print(" => Non-Stationary\n")
    print('--------------------------------------------')

# Run on renamed columns
adf_test(df['repo_rate'], 'Repo Rate')
adf_test(df['cpi_inflation'], 'CPI Inflation')
adf_test(df['wpi_inflation'], 'WPI Inflation')
adf_test(df['gdp_growth_rate'], 'GDP Growth Rate')
adf_test(df['exchange_rate'], 'Exchange Rate (INR/USD)')
adf_test(df['m3'], 'Broad Money M3')
adf_test(df['fiscal_deficit'], 'Fiscal Deficit')
adf_test(df['oil_price'], 'Oil Price')
adf_test(df['industrial_production'], 'Industrial Production')
adf_test(df['export_rate'], 'Export Rate')
adf_test(df['import_rate'], 'Import Rate')
```

###### Non-Stationary Variables (p > 0.05) are 
Repo Rate,CPI Inflation,WPI Inflation,GDP Growth Rate,Exchange Rate,M3,Fiscal Deficit,Oil Price  ......8 out of 11 variables are NON-stationary.

```python
#### Transform Non-Stationary Variables
# First differences
df['d_repo_rate'] = df['repo_rate'].diff()
df['d_cpi_inflation'] = df['cpi_inflation'].diff()
df['d_wpi_inflation'] = df['wpi_inflation'].diff()
df['d_gdp_growth_rate'] = df['gdp_growth_rate'].diff()
df['d_fiscal_deficit'] = df['fiscal_deficit'].diff()

# Log differences (growth rates)
df['dlog_exchange_rate'] = np.log(df['exchange_rate']).diff()
df['dlog_m3'] = np.log(df['m3']).diff()
df['dlog_oil_price'] = np.log(df['oil_price']).diff()

# Stationary variables (keep as they are)
# industrial_production
# export_rate
# import_rate
```
